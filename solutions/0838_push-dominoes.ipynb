[838. Push Dominoes](https://leetcode.com/problems/push-dominoes/)

最初用`'.L.R...LR..L..'`的形式表示要推哪些骨牌, 返回最终状态.
- 如果一个牌左侧是`'R'`, 右侧是`'L'`, 那么还是保持直立. 如`'R.L' -> 'R.L'`.
- 如果左牌`'R'`, 右牌是`'L'`, 是OK的, 如`'R..L' -> 'RRLL'`
- 同一方向的骨牌对力的贡献不会累计, 如`'RR.L' -> 'RR.L'`

```python
class Solution:
    def pushDominoes(self, dominoes: str) -> str:
        symbols = [(idx, type_) for idx, type_ in enumerate(dominoes) if type_ != '.']
        symbols = [(-1, 'L')] + symbols + [(len(dominoes), 'R')]
        
        res = list(dominoes)
        for (i, x), (j, y) in zip(symbols, symbols[1:]):
            if x == y:
                for k in range(i + 1, j):
                    res[k] = x
            elif x == 'R' and y == 'L':
                m = (i + j) // 2
                
                # R..L
                if (j - i) % 2:
                    for k in range(i + 1, m + 1):
                        res[k] = x
                    for k in range(m + 1, j):
                        res[k] = y
                # R...L
                else:
                    for k in range(i + 1, m):
                        res[k] = x
                    for k in range(m + 1, j):
                        res[k] = y
        
        return ''.join(res)
```

```python
class Solution:
    def pushDominoes(self, dominoes: str) -> str:
        dominoes = list(dominoes)
        N = len(dominoes)
        while True:
            processed = set()
            # 正向处理一次L
            for i in range(1, N):
                if dominoes[i] == 'L' and dominoes[i-1] == '.':
                    dominoes[i-1] = 'L'
                    processed.add(i-1)
            
            # 反向处理一次R, 此时需要考虑是否要复原L的处理
            for i in range(N - 1, -1, -1):
                if dominoes[i] == 'R':
                    if dominoes[i+1] == '.':
                        if i+2 not in processed:
                            dominoes[i+1] = 'R'
                            processed.add(i+1)
                    elif dominoes[i+1] == 'L':
                        if i+1 in processed:
                            dominoes[i+1] = '.'
                            processed.add()
                    else:
                    if i+1 in processed:
                        
                    dominoes[i+1] = {
                        '.': 'R',
                        'R': 'R',
                        'L': 'L'
                    }[dominoes[i+1]]
                    
            
            if not processed:
                break
        
        return ''.join(dominoes)
```
