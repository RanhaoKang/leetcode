[50. Pow(x, n)](https://leetcode.com/problems/powx-n/)

Top-Down with memo, 44ms

```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        memo = {
            0: 1,
            1: x,
            2: x * x,
            -1: 1 / x,
            -2: 1 / x / x
        }
        
        def solve(n):
            if n in memo:
                return memo[n]
            else:
                res1 = solve(n // 2)
                res2 = solve(n - n // 2)
                memo[n // 2] = res1
                memo[n - n // 2] = res2
                return res1 * res2
        
        return solve(n)
            
```
