[51. N-Queens](https://leetcode.com/problems/n-queens/)

backtracking, 在`n=9`时TLE

```python
class Solution:
    def solveNQueens(self, n: int):
        board = [['.'] * n for _ in range(n)]
        
        def valid(I, J):
            if any(board[I][j] == 'Q' for j in range(J)) or\
                any(board[i][J] == 'Q' for i in range(I)) or\
                any(board[i][j] == 'Q' for i, j in zip(range(I, -1, -1), range(J, -1, -1)) if i >= 0 and j >= 0) or\
                any(board[i][j] == 'Q' for i, j in zip(range(I, -1, -1), range(J, n)) if i >= 0):
                return False

            return True
        
        def solve(idx, rest):
            for idx in range(idx, n * n):
                i, j = idx // n, idx % n
                if valid(i, j):
                    board[i][j] = 'Q'
                    if rest == 1:
                        yield [''.join(i) for i in board]
                    yield from solve(idx + 1, rest - 1)
                board[i][j] = '.'
                
        return list(solve(0, n))
```

```python
Solution().solveNQueens(4)
```

```python
class Solution:
    def solveNQueens(self, n: int):
        def DFS(queens, xy_dif, xy_sum):
            p = len(queens)
            if p==n:
                result.append(queens)
                return None
            for q in range(n):
                if q not in queens and p-q not in xy_dif and p+q not in xy_sum: 
                    DFS(queens+[q], xy_dif+[p-q], xy_sum+[p+q])  
        result = []
        DFS([],[],[])
        return [ ["."*i + "Q" + "."*(n-i-1) for i in sol] for sol in result]
```
