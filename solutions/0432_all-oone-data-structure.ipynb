[432. All O`one Data Structure](https://leetcode.com/problems/all-oone-data-structure/)

```python
class Block:
    def __init__(self, val = 0):
        self.val = val
        self.keys = set()
        self.before = None
        self.after = None

    def remove(self):
        self.before.after = self.after
        self.after.before = self.before
        self.before, self.after = None, None

    def insert_after(self, new_block):
        old_after = self.after
        self.after = new_block
        new_block.before = self
        new_block.after = old_after
        old_after.before = new_block

class AllOne:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.head, self.tail = Block(), Block()
        self.head.after, self.tail.before = self.tail, self.head
        self.key_to_block = {}


    def inc(self, key: str) -> None:
        """
        Inserts a new key <Key> with value 1. Or increments an existing key by 1.
        """
        if key not in self.key_to_block:
            block = self.head
        else:
            block = self.key_to_block[key]
            block.keys.remove(key)
        
        if block.val + 1 != block.after.val:  # insert new block
            new_block = Block(block.val + 1)
            block.insert_after(new_block)
        else:
            new_block = block.after

        new_block.keys.add(key)  # update new_block
        self.key_to_block[key] = new_block  # ... and mapping of key to new_block

        if not block.keys and block.val != 0:  # delete current block if not seninel
            block.remove()


    def dec(self, key: str) -> None:
        """
        Decrements an existing key by 1. If Key's value is 1, remove it from the data structure.
        """
        if not key in self.key_to_block:
            return

        current_block = self.key_to_block[key]
        del self.key_to_block[key]  # could use self.mapping.pop(key)
        current_block.keys.remove(key)

        if current_block.val != 1:
            if current_block.val - 1 != current_block.before.val:  # insert new block
                new_block = Block(current_block.val - 1)
                current_block.before.insert_after(new_block)
            else:
                new_block = current_block.before
            new_block.keys.add(key)
            self.key_to_block[key] = new_block

        if not current_block.keys:  # delete current block
            current_block.remove()
        

    def getMaxKey(self):
        if self.tail.before.val == 0:
            return ""
        key = self.tail.before.keys.pop()  # pop and add back to get arbitrary (but not random) element
        self.tail.before.keys.add(key)

        return key


    def getMinKey(self):
        if self.head.after.val == 0:
            return ""
        key = self.head.after.keys.pop()
        self.head.after.keys.add(key)

        return key


# Your AllOne object will be instantiated and called as such:
# obj = AllOne()
# obj.inc(key)
# obj.dec(key)
# param_3 = obj.getMaxKey()
# param_4 = obj.getMinKey()
```
